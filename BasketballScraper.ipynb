```python
dfs = []
for year in years:
    file_path = os.path.join(mvp_dir, f"{year}.html")
    
    with open(file_path, 'r', encoding='utf-8') as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="over_header").decompose()  # Remove the over_header row
    
    mvp_table = soup.find(id="mvp")
    
    if mvp_table:
        mvp_df = pd.read_html(str(mvp_table))[0]
        mvp_df["Year"] = year
        dfs.append(mvp_df)
    else:
        print(f"Warning: No MVP table found for year {year}")

# dfs
```

```python
mvps = pd.concat(dfs)
mvps.head()
# mvps.tail()
```

```python
# Now process the saved HTML files and convert to CSV
dfs = []

for year in years:
    file_path = os.path.join(player_dir, f"{year}.html")
    
    with open(file_path, 'r', encoding='utf-8') as f:
        page = f.read()
    
    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(page, 'html.parser')
    # thead_tag = soup.find('thead')
    # print(f"thead_tag = {thead_tag}")
    # header = thead_tag.find('tr')
    # print(f"header = {header}")
    # break
    
    
    # Find the player stats table
    player_table = soup.find(id="per_game_stats")

    # Remove the no rank row
    player_table.find('tr', class_="norank").decompose()
    
    if player_table:
        # Convert the table into a DataFrame
        player_df = pd.read_html(str(player_table))[0]
        
        # Add the year column to the DataFrame
        player_df["Year"] = year
        
        # Append the DataFrame to the list of DataFrames
        dfs.append(player_df)
    else:
        print(f"Warning: No player table found for year {year}")
```

```python
# Combine all DataFrames into one
players = pd.concat(dfs)
players.head()
```

```python
# Save the final DataFrame to a CSV file
players.to_csv("players.csv", index=False)
print("Player CSV creation complete!")
```

```python
dfs = []
for year in years:
    file_path = os.path.join(team_dir, f"{year}.html")
    
    with open(file_path, 'r', encoding='utf-8') as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()  # Remove the over_header row
    
    e_table = soup.find(id="divs_standings_E")
    
    if e_table:
        e_df = pd.read_html(str(e_table))[0]
        e_df["Year"] = year
        e_df["Team"] = e_df["Eastern Conference"]
        del e_df["Eastern Conference"]
        dfs.append(e_df)
    
    w_table = soup.find(id="divs_standings_W")
    
    if w_table:
        w_df = pd.read_html(str(w_table))[0]
        w_df["Year"] = year
        w_df["Team"] = w_df["Western Conference"]
        del w_df["Western Conference"]
        dfs.append(w_df)
    
    else:
        print(f"Warning: No Team table found for year {year}")

# dfs
```

```python
teams = pd.concat(dfs)
# teams.head()
teams
```

```python
# Save the final DataFrame to a CSV file
teams.to_csv("team.csv", index=False)
print("Team CSV creation complete!")
```
