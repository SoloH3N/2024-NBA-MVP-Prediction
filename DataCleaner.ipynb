```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
mvps.head()
```

```python
players = pd.read_csv("players.csv")
players
```

```python
# Delete the Rank column
del players["Rk"]
players.head()
```

```python
players["Player"].head(50)
```

```python
# Replace the "*" with "" in the player column
players["Player"] = players["Player"].str.replace("*", "", regex=False)
# players["Player"].head(50)
```

```python
players.head(20)
```

```python
# Get the player and the year using groupby function
# players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
```

```python

# Function to only get a single row of the player per year
players.groupby(["Player", "Year"])
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Team"] == "3TM"]
        row["Team"] = df.iloc[-1,:]["Team"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
# Drop the first 2 columns
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
# Merge the 2 data frames of MVPs and Players
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
combined.head(5)
# "Pts Won", "Pts Max" and "Share" should be NaN if they were not in MVP data
```

```python
# Check if the "Pts Won", "Pts Max" and "Share" is correctly combined
combined[combined["Pts Won"] > 0]
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
combined[combined["Pts Won"] > 0]
```

```python
combined
```

```python
teams = pd.read_csv("team.csv")
teams
```

```python
teams.head(30)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
teams.head(30)
```

```python
teams.loc[:, "Team"] = teams["Team"].str.replace("*", "", regex=False)
teams.head(5)
```

```python
nicknames = {}

with open("nicknames.txt", "r") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name, = line.replace("\n", "").split(",")
        nicknames[abbrev] = name

nicknames
```

```python
combined["Team"] = combined["Team"].map(nicknames)
combined.head(5)
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
stats
```

```python
# Check if the number of rows is the same after the merge
# combined
```

```python
stats.dtypes
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
# Replace "—" to "0"
stats["GB"] = stats["GB"].str.replace("—", "0")
stats["GB"].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```

```python
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(20)
highest_scoring
```

```python
highest_scoring.plot.bar("Player", "PTS")
```
